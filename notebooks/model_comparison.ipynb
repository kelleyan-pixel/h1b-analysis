```python
def build_small_X(data):
    n, K, N_S, N_ST = len(data["y"]), data["K"], data["N_S"], data["N_ST"]
    cols = [np.ones((n, 1))]
    for k in range(1, K):
        cols.append((data["L"] == k).reshape(-1, 1).astype(np.float64))
    for s in range(1, N_S):
        cols.append((data["S"] == s).reshape(-1, 1).astype(np.float64))
    for st in range(1, N_ST):
        cols.append((data["ST"] == st).reshape(-1, 1).astype(np.float64))
    return np.ascontiguousarray(np.hstack(cols))


# =========================
# HELPERS
# =========================

def predict_classical(result, data):
    X  = build_small_X(data)
    mu = X @ result["beta"] + result["alpha"][data["E"]]
    return mu, float(result["sigma"])

def predict_latent(result, data):
    mu = (result["beta0"]
          + result["gamma"][data["L"]]
          + result["alpha"][data["E"]]
          + result["delta"][data["S"]]
          + result["eta"][data["ST"]])
    return mu, float(result["sigma"])

def tobit_loglik(y, censored, mu, sigma):
    obs  = ~censored
    cens =  censored
    ll = (stats.norm.logpdf((y[obs] - mu[obs]) / sigma) - np.log(sigma)).sum() \
       + log_ndtr(-mu[cens] / sigma).sum()
    return float(ll)


# =========================
# 1. IN-SAMPLE FIT — AIC / BIC
# =========================

def count_params_classical(result):
    d = result["data"]
    return 1 + 3 + (d["N_S"] - 1) + (d["N_ST"] - 1) + d["N_E"] + 1

def count_params_latent(result):
    d = result["data"]
    return 1 + d["K"] + d["N_E"] + d["N_S"] + d["N_ST"] + 1 + 4

def insample_comparison(result_c, result_l):
    data     = result_c["data"]
    y        = data["y"]
    censored = data["censored"]
    n        = len(y)

    mu_c, sig_c = predict_classical(result_c, data)
    mu_l, sig_l = predict_latent(result_l,   data)

    ll_c = tobit_loglik(y, censored, mu_c, sig_c)
    ll_l = tobit_loglik(y, censored, mu_l, sig_l)
    k_c  = count_params_classical(result_c)
    k_l  = count_params_latent(result_l)

    rows = []
    for name, ll, k in [("Classical MLE", ll_c, k_c), ("Latent EM", ll_l, k_l)]:
        rows.append({
            "Model":   name,
            "log-lik": round(ll, 1),
            "params":  k,
            "AIC":     round(-2 * ll + 2 * k, 1),
            "BIC":     round(-2 * ll + k * np.log(n), 1),
        })

    df_out = pd.DataFrame(rows).set_index("Model")
    print("\n=== In-sample fit ===")
    print(df_out.to_string())
    return df_out


# =========================
# 2. HELD-OUT PREDICTIVE ACCURACY — 80/20 split
# Requires fit_model_c and fit_model_l to be defined in the kernel
# from running the other two notebooks first.
# =========================

def held_out_comparison(df_full, prepare_data_fn_c, prepare_data_fn_l,
                         fit_fn_c, fit_fn_l, seed=42, test_frac=0.2):
    rng       = np.random.default_rng(seed)
    idx       = np.arange(len(df_full))
    test_idx  = rng.choice(idx, size=int(len(idx) * test_frac), replace=False)
    train_idx = np.setdiff1d(idx, test_idx)

    df_train = df_full.iloc[train_idx].reset_index(drop=True)
    df_test  = df_full.iloc[test_idx].reset_index(drop=True)

    print("Fitting classical on train …")
    res_c = fit_fn_c(prepare_data_fn_c(df_train), verbose=False)
    print("Fitting latent EM on train …")
    res_l = fit_fn_l(prepare_data_fn_l(df_train))

    # restrict test to entities seen in training
    known_E  = set(df_train["EMPLOYER_NAME_CLEAN"].unique())
    known_ST = set(df_train["WORKSITE_STATE"].unique())
    df_test  = df_test[
        df_test["EMPLOYER_NAME_CLEAN"].isin(known_E) &
        df_test["WORKSITE_STATE"].isin(known_ST)
    ].reset_index(drop=True)

    data_tc = prepare_data_fn_c(df_test)
    data_tl = prepare_data_fn_l(df_test)

    mu_c, sig_c = predict_classical(res_c, data_tc)
    mu_l, sig_l = predict_latent(res_l,   data_tl)

    y_test = data_tc["y"]
    obs    = ~data_tc["censored"]

    rows = []
    for name, mu, sig in [("Classical MLE", mu_c, sig_c), ("Latent EM", mu_l, sig_l)]:
        rmse = float(np.sqrt(np.mean((y_test[obs] - mu[obs]) ** 2)))
        mae  = float(np.mean(np.abs(y_test[obs] - mu[obs])))
        ls   = float(stats.norm.logpdf(y_test[obs], mu[obs], sig).mean())
        rows.append({"Model": name, "RMSE": round(rmse,5),
                     "MAE": round(mae,5), "LogScore": round(ls,4)})

    df_out = pd.DataFrame(rows).set_index("Model")
    print("\n=== Held-out accuracy (uncensored test obs) ===")
    print(df_out.to_string())
    return df_out


# =========================
# 3. CENSORING CALIBRATION by wage level
# =========================

def calibration_plot(result_c, result_l):
    data     = result_c["data"]
    censored = data["censored"]
    L        = data["L"]

    mu_c, sig_c = predict_classical(result_c, data)
    mu_l, sig_l = predict_latent(result_l,   data)

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    level_labels = ["I", "II", "III", "IV"]

    for ax, (name, mu, sig) in zip(axes, [
        ("Classical MLE", mu_c, sig_c),
        ("Latent EM",     mu_l, sig_l),
    ]):
        actual = [censored[L == k].mean()        for k in range(4)]
        pred   = [ndtr(-mu[L == k].mean() / sig) for k in range(4)]
        x = np.arange(4)
        ax.bar(x - 0.2, actual, 0.4, label="Actual")
        ax.bar(x + 0.2, pred,   0.4, label="Predicted P(cens)")
        ax.set_xticks(x); ax.set_xticklabels(level_labels)
        ax.set_ylabel("P(censored)")
        ax.set_title(f"{name} — Censoring Calibration")
        ax.legend()

    plt.tight_layout()
    plt.show()


# =========================
# 4. EMPLOYER EFFECT STABILITY — split-half correlation
# =========================

def employer_stability(df_full, prepare_data_fn_c, prepare_data_fn_l,
                        fit_fn_c, fit_fn_l, min_obs=50, seed=0):
    rng     = np.random.default_rng(seed)
    counts  = df_full["EMPLOYER_NAME_CLEAN"].value_counts()
    big_emp = counts[counts >= min_obs].index
    df_big  = df_full[df_full["EMPLOYER_NAME_CLEAN"].isin(big_emp)].copy()

    half_A, half_B = [], []
    for _, grp in df_big.groupby("EMPLOYER_NAME_CLEAN"):
        idx = grp.index.tolist()
        rng.shuffle(idx)
        mid = len(idx) // 2
        half_A.extend(idx[:mid])
        half_B.extend(idx[mid:])

    df_A = df_full.loc[half_A].reset_index(drop=True)
    df_B = df_full.loc[half_B].reset_index(drop=True)

    print("Fitting on split A …")
    res_cA = fit_fn_c(prepare_data_fn_c(df_A), verbose=False)
    res_lA = fit_fn_l(prepare_data_fn_l(df_A))
    print("Fitting on split B …")
    res_cB = fit_fn_c(prepare_data_fn_c(df_B), verbose=False)
    res_lB = fit_fn_l(prepare_data_fn_l(df_B))

    cats_A = pd.Categorical(df_A["EMPLOYER_NAME_CLEAN"])
    cats_B = pd.Categorical(df_B["EMPLOYER_NAME_CLEAN"])
    shared = sorted(set(cats_A.categories) & set(cats_B.categories))
    iA = {e: i for i, e in enumerate(cats_A.categories)}
    iB = {e: i for i, e in enumerate(cats_B.categories)}

    acA = res_cA["alpha"][[iA[e] for e in shared]]
    acB = res_cB["alpha"][[iB[e] for e in shared]]
    alA = res_lA["alpha"][[iA[e] for e in shared]]
    alB = res_lB["alpha"][[iB[e] for e in shared]]

    r_c = float(np.corrcoef(acA, acB)[0, 1])
    r_l = float(np.corrcoef(alA, alB)[0, 1])

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    for ax, (name, aA, aB, r) in zip(axes, [
        ("Classical MLE", acA, acB, r_c),
        ("Latent EM",     alA, alB, r_l),
    ]):
        ax.scatter(aA, aB, alpha=0.3, s=10)
        lims = [min(aA.min(), aB.min()), max(aA.max(), aB.max())]
        ax.plot(lims, lims, "r--", lw=1)
        ax.set_xlabel("Half A"); ax.set_ylabel("Half B")
        ax.set_title(f"{name} employer FE stability\nr = {r:.3f}")

    plt.tight_layout()
    plt.show()
    print(f"\nSplit-half r — Classical: {r_c:.3f}  |  Latent EM: {r_l:.3f}")
    return r_c, r_l


# =========================
# 5. SIDE-BY-SIDE DIAGNOSTIC PLOT
# =========================

def summary_plot(result_c, result_l):
    data     = result_c["data"]
    y        = data["y"]
    obs      = ~data["censored"]

    mu_c, sig_c = predict_classical(result_c, data)
    mu_l, sig_l = predict_latent(result_l,   data)

    resid_c = (y[obs] - mu_c[obs]) / sig_c
    resid_l = (y[obs] - mu_l[obs]) / sig_l

    fig, axes = plt.subplots(2, 3, figsize=(15, 8))
    for row, (name, resid, sig, alpha) in enumerate([
        ("Classical MLE", resid_c, sig_c, result_c["alpha"]),
        ("Latent EM",     resid_l, sig_l, result_l["alpha"]),
    ]):
        stats.probplot(resid, plot=axes[row, 0])
        axes[row, 0].set_title(f"{name} — Q-Q")

        ax = axes[row, 1]
        ax.hist(resid, bins=60, density=True, alpha=0.7)
        xr = np.linspace(-5, 5, 200)
        ax.plot(xr, stats.norm.pdf(xr), "r-", lw=1.5)
        ax.set_xlim(-5, 5)
        ax.set_title(f"{name} — Residuals  σ={sig:.4f}")

        axes[row, 2].hist(alpha, bins=40)
        axes[row, 2].set_title(f"{name} — Employer FEs")

    plt.tight_layout()
    plt.show()


# =========================
# RUN — fast tests (no refitting)
# =========================
summary_plot(result_c, result_l)
insample_comparison(result_c, result_l)
calibration_plot(result_c, result_l)

# =========================
# RUN — slow tests (refit on subsets, uncomment when ready)
# These need the prepare_data and fit_model functions from the
# other notebooks to be defined in this kernel session.
# =========================
# held_out_comparison(df, prepare_data, prepare_data, fit_model_c, fit_model_l)
# employer_stability(df, prepare_data, prepare_data, fit_model_c, fit_model_l)
```
